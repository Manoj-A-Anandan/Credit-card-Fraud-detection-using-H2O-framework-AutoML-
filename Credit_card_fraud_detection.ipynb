```python
!pip install h2o
import h2o
from h2o.automl import H2OAutoML

h2o.init()
```

```python


# Load data into H2O
data = h2o.import_file("/content/credit card and fraud detection.csv")

# Define features (X) and target (y)
X = data.columns[:-1]  # All columns except the last one (target)
y = data.columns[-1]   # Last column (target)

# Convert target column to categorical
data[y] = data[y].asfactor()

# Initialize AutoML
aml = H2OAutoML(max_runtime_secs=60)  # Example: maximum runtime of 1 hour

# Train AutoML
aml.train(x=X, y=y, training_frame=data)

# View AutoML leaderboard
lb = aml.leaderboard
print(lb)

# Optionally, select the best model for further evaluation or deployment
best_model = aml.leader
print(best_model)

# Shut down H2O cluster
h2o.cluster().shutdown()
```
